# Random Forest - regression --->  PHASE 1 model creation. 
## PROBLEM STATEMENT: 
### Profit prediction for startups based on the investment rate in different departments

## dataset, train and test split

```python
# read the data set
import pandas as pd
dataset = pd.read_csv("50_Startups.csv")
dataset.head()
```

```python
dataset.columns
```

```python
# convert categorial into numerical data
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset.head()
```

```python
# split i/p o/p
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York' ]]
independent.head()
```

```python
dependent = dataset[[ 'Profit']] 
dataset.head()
```

```python
# split train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size=0.20, random_state=0)
```

## model creation

```python
# model creation 
"""
criterion --- {'squared_error', 'absolute_error', 'poisson', 'friedman_mse'}
"""

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=0)  #n_estimator -- no of trees in the forest  (hyper parameter)
regressor.fit(x_train,y_train)
```

```python
# test the model 
y_predicted = regressor.predict(x_test)
y_predicted
```

```python
# evaluation metrics
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_predicted)
r_score    #0.9700374289598385 -- best accuracy
```

## Save the model

```python
# save the model
import pickle 
pickle.dump(regressor, open('finalized_model_RandomForestRegressor.sav' ,'wb'))
```

```python
# load the model
loaded_model = pickle.load(open('finalized_model_RandomForestRegressor.sav','rb'))
```

```python
# predict the model
loaded_model.predict([[10000, 5000, 20000,0,1]])   # input data :   'R&D Spend', 'Administration', 'Marketing Spend',  'State_Florida', 'State_New York' 
```
